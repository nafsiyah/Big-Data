# Data Cleaning

Proyek ini menganilisis angka harapan hidup dan faktor yang memengaruhinya. Data yang dianalisis adalah data tahun 2015. Karena data yang diperoleh adalah data tahun 2000 sampai 2015 maka data tahun 2000-2014 dihapus dari DataFrame.

```python
# Melakukan filter data tahun 2015

data2 = data[data['Year']==2015] 

data2.head(20)
```

Dari hasil cleaning, diperoleh data hanya tahun 2015 seperti yang terlihat pada tabel di atas. Dari tabel tersebut terlihat bahwa banyak data pada kolom "percentage expenditure" dan "Total expenditure" yang kosong. Sehingga pada proyek ini, kolo, tersebut dihapus.

```python
# Menghapus kolom 'percentage expenditure' dan 'Total expenditure'

data = data2.drop(['percentage expenditure', 'Total expenditure'], axis=1)
data
```

Hasil penghapusan kolom terlihat pada tabel di atas. Dari tabel tersebut terlihat bahwa terdapatdata yang kosong sehingga dilakukan penghapusan data yang kosong agar mempermudah proses pengolahan data.

```python
# Penghapusan Data yang Kosong
data = data.dropna()
data
```

Proses Cleaning data selesai dan diperoleh data yang akan dianalisis seperti yang terlihat pada tabel di atas

# MODEL REGRESI

Berdasarkan scatter plot yang telah diperoleh, diketahui bahwa kedua pola mempunya pola yang linier sehinggamodel yang sesuai adalah model regresi linier.

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn import datasets, linear_model, metrics 
from sklearn.linear_model import LinearRegression
```

## 1. Model Regresi antara Income composition of resources dan Life Expectancy

```python
x1 = data['Income composition of resources']
y1 = data['Life expectancy']

x1 = np.array(x1).reshape((-1,1))
y1 = np.array(y1)


model1 = LinearRegression()
model1.fit(x1,y1)
model1 = LinearRegression().fit(x1, y1)

print('b_0: ', model1.intercept_)
print('b_1: ', model1.coef_)


r_sq1 = model1.score(x1, y1)
print('R^2: ', r_sq1)

print('\nModel Regresi antara Income composition of resources dan Life Expectancy')
print('y = ', model1.intercept_, '+', model1.coef_, 'x')
print('Life Expectancy = ', model1.intercept_, '+', model1.coef_, 'Income composition of resources')
```

```python
y1_pred = model1.predict(x1)
plt.scatter(x1,y1)
plt.plot(x1,y1_pred, color = 'red', lw = 5)
plt.title('Scatter plot dan Hasil Regresi antara Income composition of resources dan Life Expectancy')
plt.xlabel('Income composition of resources')
plt.ylabel('Life Expectancy')
plt.show()
```

### Tingkat Keakuratan

```python
target = y1
prediction = y1_pred
```

```python
# mean absolute error
mae = mean_absolute_error(target, prediction)
print("MAE:", mae)

# mean squared error
mse = mean_squared_error(target, prediction)
print("MSE:", mse)

# r2
r2 = r2_score(target, prediction)
print("R2:", r2)
```

```python
from pandas.plotting import scatter_matrix
from matplotlib import cm
feature_names = ['Adult Mortality', 'BMI', 'GDP', 'Income composition of resources', 'Life expectancy',
                                         'Population', 'Schooling', 'infant deaths', 'under-five deaths']
X = data[feature_names]
#y = data['Status'].astype('category').cat.codes
#y = data['Status']
cmap = cm.get_cmap('gnuplot')
scatter = scatter_matrix(X, marker = 'o', s=40, hist_kwds={'bins':15}, figsize=(15,15), cmap = cmap)
plt.suptitle('Scatter-matrix for each input variable')
plt.savefig('Status_scatter_matrix')
```

```python
from sklearn.model_selection import train_test_split
y = data['Status']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier().fit(X_train, y_train)
print('Accuracy of Decision Tree classifier on training set: {:.2f}'
     .format(clf.score(X_train, y_train)))
print('Accuracy of Decision Tree classifier on test set: {:.2f}'
     .format(clf.score(X_test, y_test)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test, y_test)))
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
print('Accuracy of Logistic regression classifier on training set: {:.2f}'
     .format(logreg.score(X_train, y_train)))
print('Accuracy of Logistic regression classifier on test set: {:.2f}'
     .format(logreg.score(X_test, y_test)))
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
print('Accuracy of LDA classifier on training set: {:.2f}'
     .format(lda.score(X_train, y_train)))
print('Accuracy of LDA classifier on test set: {:.2f}'
     .format(lda.score(X_test, y_test)))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
print('Accuracy of GNB classifier on training set: {:.2f}'
     .format(gnb.score(X_train, y_train)))
print('Accuracy of GNB classifier on test set: {:.2f}'
     .format(gnb.score(X_test, y_test)))
```
